# Pendulum swing

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_theta_2dot(theta, theta_dot):
    # definition of ODE
    return -mu * theta_dot - (g/L) * np.sin(theta)

def theta(t, timestep, theta_0, theta_dot_0): # having time array helps avoid confusion of x-y axes array dimension mismatch 
    theta = theta_0
    theta_dot = theta_dot_0
    for time in np.arange(0,t,timestep): 
        theta_2dot = get_theta_2dot(theta, theta_dot)
        theta_dot += theta_2dot * timestep
        theta += theta_dot * timestep
    return theta

```

```python
# Parameters
mu = 0.7 # drag constnant 
L = 2 # length of string [m]
g = 9.8 # Gravitational acceleration [m/s^2]

# Initial conditions
theta_0 = np.pi / 3 # starting angle
theta_dot_0 = 0 # starting angular velocity
timestep = 0.01


t = np.arange(0,10,timestep)
Theta = [theta(i, timestep, theta_0, theta_dot_0) for i in t]

plt.plot(t,Theta)
plt.xlabel("Time")
plt.ylabel("Angle [rad]")
plt.grid()
```

# Free fall

```python
def get_x2dot(xdot):
    return g - (k/m)* xdot # K = (k/m)

def x(t, timestep):
    x = x0
    xdot = xdot0
    for time in np.arange(0,t, timestep):
        x2dot = -get_x2dot(xdot)
        xdot += x2dot*timestep
        x += xdot*timestep
    return x
```

```python
x0 = 500 # [m]
xdot0 = 0 # [m/s]
g = 9.8 #[m/s2]
k = 0.1
m = 100 #[kg]
timestep = 0.01 #[s]
t_n = 10 # [s]

t = np.arange(0,t_n,timestep)
xs = [x(i, timestep) for i in t]

plt.figure()
plt.plot(t,xs)
plt.ylabel("position")
plt.xlabel("time")
plt.grid()
```
